# Question 1 ----Plot male/female distribution in Pie-Chart

```python
x=titanic['sex'].value_counts()
x
```

```python
plt.pie(x, explode=None, labels=('Male','Female') ,autopct='%1.1f%%', shadow=True)
plt.show()
```

# Question 2: Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
x1=titanic['age']
y=titanic['fare']
colors=('g','r')
fig,ax=plt.subplots()
ax.scatter(x1, y,c=colors, label=x)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Plot Fare vs Age by Gender')
plt.legend(loc=2)
```
